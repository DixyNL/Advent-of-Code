```python
with open("day3.txt", "r") as file:
    rows = file.read().strip().split('\n')

grid = [list(row) for row in rows]
numbs = []
def find_sequences(grid):
    sequences = []
    num_rows, num_cols = len(grid), len(grid[0])

    for row_index in range(num_rows):
        for col_index in range(num_cols):
            if grid[row_index][col_index].isdigit():
                current_sequence = [int(grid[row_index][col_index])]
                current_position = [row_index, col_index]

                for offset in range(1, 10): 
                    new_row, new_col = row_index, col_index + offset
                    if 0 <= new_col < num_cols and grid[new_row][new_col].isdigit():
                        current_sequence.append(int(grid[new_row][new_col]))
                    else:
                        break

                if len(current_sequence) >= 1:
                    value = int("".join(map(str, current_sequence)))
                    end_position = [row_index, col_index + len(current_sequence) - 1]
                    numbs.append([value, [current_position, end_position]])
                    sequences.append({"value": value, "positions": [current_position, end_position]})
                    
                    for pos in range(col_index, col_index + len(current_sequence)):
                        grid[row_index][pos] = 'x'

    return sequences

all_syms = ["*", "$", "/", "@", "#", "%", "&", "*", "+", "=", "-"]
part2 = ["*"]
def isSymbol(char, symbol_array):
    return char in symbol_array

def find_symbols(grid, symbol_array):
    symbols = []
    num_row , numb_colmn = len(grid), len(grid[0])
    
    for row_index in range(num_row):
        for col_index in range(numb_colmn):
            if isSymbol(grid[row_index][col_index], symbol_array):
                symbols.append({"symbol": grid[row_index][col_index], "positions": [row_index, col_index]})
    return symbols

consecutive_digit_sequences = find_sequences(grid)
find_symbol_position = find_symbols(grid, all_syms)

symb_pos = []
for symbs in find_symbol_position:
    symb_pos.append([symbs['positions']])
```

```python
def is_adjacent(pos1, pos2):
    # Helper function to check if two positions are adjacent
    return abs(pos1[0] - pos2[0]) <= 1 and abs(pos1[1] - pos2[1]) <= 1

def find_adjacent_numbers(symbol_pos, numbs):
    adjacent_numbers = []

    for symbol in symbol_pos:
        symbol_position = symbol[0]

        for number in numbs:
            num_position_1, num_position_2 = number[1]

            if is_adjacent(symbol_position, num_position_1) or is_adjacent(symbol_position, num_position_2):
                adjacent_numbers.append(number[0])

    return adjacent_numbers

adjacent_numbers = find_adjacent_numbers(symb_pos, numbs)
print(adjacent_numbers)
```
